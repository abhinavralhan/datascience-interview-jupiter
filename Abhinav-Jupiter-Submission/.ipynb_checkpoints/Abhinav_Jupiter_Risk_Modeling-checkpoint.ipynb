### Feature cleaning / engineering

```python
# Dropping columns with less than 25% fill
```

```python
fill_rate = df.isna().sum().reset_index().rename(columns={'index': 'columns', 0: 'fill_rate'})
fill_rate['fill_rate'] = (100 * fill_rate['fill_rate'] / len(df))
fill_rate = fill_rate[fill_rate.fill_rate < 75]
df = df.iloc[:, df.columns.isin(fill_rate['columns'].tolist())]
```

```python
df.shape
```

```python
# Fixing some columns
```

```python
# Emp_title cleanup
```

```python
df.emp_title = df.emp_title.str.replace(" ","")
df.emp_title = df.emp_title.str.lower()
emp_title_value_counts = df.emp_title.value_counts().reset_index()
emp_title_value_counts = emp_title_value_counts.rename(columns={'emp_title': 'emp_title_counts', 'index': 'emp_title'})
others = emp_title_value_counts[emp_title_value_counts.emp_title_counts < 50].emp_title.tolist()
df.loc[df['emp_title'].isin(others), 'emp_title'] = 'other'
```

```python
# loan title
```

```python
df.title = df.title.str.replace(" ","")
df.title = df.title.str.lower()
title_value_counts = df.title.value_counts().reset_index().rename(columns={'title': 'title_count', 'index': 'title'})
others = title_value_counts.loc[title_value_counts.title_count < 50].title.tolist()
df.loc[df.title.isin(others), 'title'] = 'other'
```

```python
# zip code
```

```python
zip_code_value_counts = df['zip_code'].value_counts().reset_index().rename(columns={'zip_code': 'zip_code_count', 'index': 'zip_code'})
others = zip_code_value_counts.loc[zip_code_value_counts['zip_code_count'] < 100].zip_code.tolist()
df.loc[df['zip_code'].isin(others), 'zip_code'] = 'other'
```

```python
# handle all date features. drop original date column and fetches recency aspect of that feature

all_date_columns = ['last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d', 'earliest_cr_line', 'issue_d']

for column in all_date_columns:
    col_value_counts = df[column].value_counts().reset_index().rename(columns={column: f'{column}_count', 'index': f'{column}_source'})
    replacements = col_value_counts[f'{column}_source'].apply(lambda x: datetime.strptime(x, '%b-%Y').strftime('%d/%m/%Y'))
    replacements = replacements.to_frame().rename(columns={f'{column}_source': f'{column}'})
    col_value_counts = col_value_counts.join(replacements)
    column_mapping = pd.Series(col_value_counts[f'{column}'].values,index=col_value_counts[f'{column}_source']).to_dict()
    df[f'{column}'] = df[f'{column}'].map(column_mapping)
    df[f'{column}'] = pd.to_datetime(df[f'{column}'])

    df[f'time_from_{column}'] = df[f'{column}'].apply(lambda x: (datetime.now() - x))
    df[f'time_from_{column}'] = df[f'time_from_{column}'].dt.days
    df = df.drop(columns=[f'{column}'])
```

# Scorecard modeling

## Assumptions being made for modeling -
> 1. Considering current and in grace period loans to be a part of "PAID" cohort loan status
> 2. Considering all late, charged off loans to be a part of "DEFAULT" state

```python
df.loc[df.loan_status.isin(['Fully Paid', 'Does not meet the credit policy. Status:Fully Paid', 'Current', 'In Grace Period']), 'loan_status'] = 0

df.loc[df.loan_status.isin(['Charged Off', 'Late (31-120 days)', 'Late (16-30 days)', 'Default', 'Does not meet the credit policy. Status:Charged Off']), 'loan_status'] = 1
```

## Prepare dataset for modeling

```python
# df.to_csv('cleaned_dataset_v3.csv')
```

```python
df = pd.read_csv('cleaned_dataset_v3.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
train, test = train_test_split(df, test_size=0.25, stratify=df['loan_status'])
```

```python
train.shape, test.shape
```

```python
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```

```python
train.loan_status.value_counts(normalize=True)
```

```python
test.loan_status.value_counts(normalize=True)
```

```python
y_train = train[['loan_status']]
X_train = train.drop(columns=['loan_status'])

y_test = test[['loan_status']]
X_test = test.drop(columns=['loan_status'])
```

```python
for column in X_train.columns:
    feature = np.array(X_train[column]).reshape(-1,1)
    scaler = MinMaxScaler()
    scaler.fit(feature)
    feature_scaled = scaler.transform(feature)
    X_train[column] = feature_scaled.reshape(1,-1)[0]

for column in X_test.columns:
    feature = np.array(X_test[column]).reshape(-1,1)
    scaler = MinMaxScaler()
    scaler.fit(feature)
    feature_scaled = scaler.transform(feature)
    X_test[column] = feature_scaled.reshape(1,-1)[0]
```

```python
lr_clf = LogisticRegression(max_iter=1000, class_weight='balanced')
lr_clf.fit(X_train, y_train)
```

```python
# save
filename = 'lr_clf_model.sav'
# filename = 'lr_clf_model_v2.sav'
# pickle.dump(lr_clf, open(filename, 'wb'))
 
# load
lr_clf = pickle.load(open(filename, 'rb'))
```

# Logistic v1

```python
valid_pred_prob = lr_clf.predict_proba(X_test)[:, 1]
valid_pred = lr_clf.predict(X_test)

print(classification_report(y_test, valid_pred))
```

```python
print("AUC:", roc_auc_score(y_test, valid_pred_prob))
print("Precision:", precision_score(y_test, valid_pred))
print("Recall:", recall_score(y_test, valid_pred))
```

```python
plot_roc_curve(lr_clf, X_test, y_test)  
plt.show()
```

# Confusion metrics related business metrics

```python
delinquent_users = test_with_scores[test_with_scores['loan_status'] == 1]
default_amnt = delinquent_users.loan_amnt.sum()
default_users = len(delinquent_users)

nondelinquent_users = test_with_scores[test_with_scores['loan_status'] == 0]
nondefault_amnt = nondelinquent_users.loan_amnt.sum()
nondefault_users = len(nondelinquent_users)

print(f"Total Default amount {default_amnt}")
print(f"Total Default users {default_users}")
print(f"Default amount/user {round(default_amnt/default_users, 3)}\n")

print(f"Total NonDefault amount {nondefault_amnt}")
print(f"Total NonDefault users {nondefault_users}")
print(f"NonDefault amount/user {round(nondefault_amnt/nondefault_users, 3)}\n")

default_users_and_flagged = test_with_scores[(test_with_scores['loan_status'] == 1) & (test_with_scores['default'] == 1)]
default_amnt_and_flagged = default_users_and_flagged.loan_amnt.sum()
default_users_and_flagged = len(default_users_and_flagged)

print(f"Total Default amount captured {default_amnt_and_flagged}, {100*default_amnt_and_flagged/default_amnt}%")
print(f"Total Default users captured {default_users_and_flagged}\n")

default_users_and_not_flagged = test_with_scores[(test_with_scores['loan_status'] == 1) & (test_with_scores['default'] == 0)]
default_amnt_and_not_flagged = default_users_and_not_flagged.loan_amnt.sum()
default_users_and_not_flagged = len(default_users_and_not_flagged)

print(f"Total Default amount not captured {default_amnt_and_not_flagged}, {100*default_amnt_and_not_flagged/default_amnt}%")
print(f"Total Default users not captured {default_users_and_not_flagged}\n")

nondefault_users_and_flagged = test_with_scores[(test_with_scores['loan_status'] == 0) & (test_with_scores['default'] == 1)]
nondefault_amnt_and_flagged = nondefault_users_and_flagged.loan_amnt.sum()
nondefault_users_and_flagged = len(nondefault_users_and_flagged)

print(f"Total NonDefault amount wrongly captured {nondefault_amnt_and_flagged}, {100*nondefault_amnt_and_flagged/nondefault_amnt}%")
print(f"Total NonDefault users wrong captured {nondefault_users_and_flagged}\n")

nondefault_users_and_notflagged = test_with_scores[(test_with_scores['loan_status'] == 0) & (test_with_scores['default'] == 0)]
nondefault_amnt_and_notflagged = nondefault_users_and_notflagged.loan_amnt.sum()
nondefault_users_and_notflagged = len(nondefault_users_and_notflagged)

print(f"Total NonDefault amount wrongly captured {nondefault_amnt_and_notflagged}, {100*nondefault_amnt_and_notflagged/nondefault_amnt}%")
print(f"Total NonDefault users wrong captured {nondefault_users_and_notflagged}")
```

# Overfitting check

```python
train_scores, test_scores = list(), list()

values = [i for i in range(1, 21)]

for i in values:
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(X_train, y_train)

    train_yhat = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    train_scores.append(train_acc)

    test_yhat = model.predict(X_test)
    test_acc = accuracy_score(y_test, test_yhat)
    test_scores.append(test_acc)

    print('>%d, train: %.3f, test: %.3f' % (i, train_acc, test_acc))
```

```python
from matplotlib import pyplot
pyplot.plot(values, train_scores, '-o', label='Train')
pyplot.plot(values, test_scores, '-o', label='Test')
pyplot.legend()
pyplot.show()
```
